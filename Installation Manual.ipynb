### Installation Manual

#### Introduction
This manual provides step-by-step instructions to set up and use the Airbnb-like database management system designed for the "Build a Data Mart in SQL" project. This database supports various functionalities, including user management, property listings, bookings, and financial transactions.

#### Prerequisites
Before proceeding with the installation, ensure you have the following software installed on your system:
- **MySQL** (version 8.0 or later)
- **DataGrip** (version 2023.1 or later)

#### Installation Steps

1. **Install MySQL**
   - **Download MySQL**: Visit the official [MySQL download page](https://dev.mysql.com/downloads/mysql/) and download the installer for your operating system.
   - **Run the Installer**: Follow the on-screen instructions to install MySQL on your machine. Ensure that you set up a root password and remember it for later use.

2. **Install DataGrip**
   - **Download DataGrip**: Visit the official [DataGrip download page](https://www.jetbrains.com/datagrip/download/) and download the installer for your operating system.
   - **Run the Installer**: Follow the on-screen instructions to install DataGrip.

3. **Set Up MySQL Server**
   - **Start MySQL Server**: Ensure the MySQL server is running. You can start the server from the MySQL Workbench or using terminal/command prompt commands:
     - On macOS or Linux:
       ```sh
       sudo service mysql start
       ```
     - On Windows:
       - Open the Services app and start the MySQL service.

4. **Create Database**
   - Open DataGrip and connect to your MySQL server using the root credentials set during installation.
   - **Create Database**:
     - In DataGrip, open a new SQL console and run the following command to create a new database:
       ```sql
       CREATE DATABASE name_of_database;
       ```

5. **Import SQL Files**
   - **Open DataGrip**: Connect to your MySQL server and select the `name_of_database` database.
   - **Run SQL Scripts**: Open the SQL files provided for the project (e.g., table creation scripts, data insertion scripts) and execute them in DataGrip. This will create the tables and populate them with initial data.
